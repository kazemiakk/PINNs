```python
import torch
from sklearn.datasets import make_blobs
from matplotlib import pyplot
from numpy import where


num_features=2
num_class=3
num_samples=1000
x,yt=make_blobs(n_samples=num_samples,n_features=num_features,centers=num_class,cluster_std=1.0, random_state=1)

for class_value in range (num_class):
  row_idx=where(yt==class_value)
  pyplot.scatter(x[row_idx,0],x[row_idx,1])

pyplot.show()

x_train=torch.tensor(x[:int(0.6*num_samples),:]).float()
y_train=torch.tensor(yt[:int(0.6*num_samples)]).long()
x_valid=torch.tensor(x[int(0.6*num_samples):int(0.75*num_samples),:]).float()
y_valid=torch.tensor(yt[int(0.6*num_samples):int(0.75*num_samples)]).long()
x_test=torch.tensor(x[int(0.75*num_samples):,:]).float()
y_test=torch.tensor(yt[int(0.75*num_samples):]).long()

num_hidden=10
model=torch.nn.Sequential(torch.nn.Linear(num_features,num_hidden),
                          torch.nn.ReLU(),
                          torch.nn.Linear(num_hidden,num_class),
                          )



loss=torch.nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(), lr=0.001)
#x=torch.FloatTensor(x)
#yt=torch.LongTensor(yt)
num_epochs=200
for epoch in range(num_epochs):
  optimizer.zero_grad()
  yp=model(x_train)
  print(yp)
  loss_value=loss(yp,y_train)
  num_corrects=torch.sum(torch.max(yp,1)[1]==y_train)
  acc_train=num_corrects.float()/num_samples.__float__()
  loss_value.backward()
  optimizer.step()

  yp=model(x_valid)
  num_corrects=torch.sum(torch.max(yp,1)[1]==y_valid)
  acc_valid=num_corrects.float()/num_samples.__float__()
  print('Epoch',epoch,'Loss:', loss_value,'accuracy_train',acc_train.item(),'accuracy_valid',acc_valid.item())

yp=model(x_test)
num_corrects=torch.sum(torch.max(yp,1)[1]==y_test)
acc_test=num_corrects.float()/num_samples.__float__()
print('Test Accuracy_valid',acc_test.item())
```
